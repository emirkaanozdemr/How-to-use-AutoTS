# How to Use AutoTS(Automatic Time Series)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import yfinance as yf
import datetime
from autots import AutoTS
```

```python
#Downloading dataset
df=yf.download("THYAO.IS",start="2016-1-1",end="2024-4-18")
```

```python
df.head()
```

```python
df.to_csv("THYAO.IS.csv")
df=pd.read_csv("THYAO.IS.csv")
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df["month"]=df.Date.dt.month
df["year"]=df.Date.dt.year
df["day"]=df.Date.dt.day
df["dayofweek"]=df.Date.dt.dayofweek
```

```python
df["dayofweek"]=df["dayofweek"].map({0:"Monday",1:"Tuesday",2:"Wednesday",3:"Thursday",4:"Friday",5:"Saturday",6:"Sunday"})
```

```python
df.tail()
```

```python
df=df[["Date","Close"]]
```

```python
model = AutoTS(forecast_length=10, frequency='infer', ensemble='simple', drop_data_older_than_periods=200)
```

```python
model = model.fit(df, date_col='Date', value_col='Close', id_col=None)
```

```python
prediction = model.predict()
forecast = prediction.forecast
print("Stock Price Prediction of Apple")
print(forecast)
```
